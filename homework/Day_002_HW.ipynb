### 如果沒有想法，可以先嘗試找出剛剛例子中提到的問題的答案
#### 資料的 row 數以及 column 數

```python
(r,c) = app_train.shape
print('Row count = %s' % (r) )
print('Column count = %s' % (c))
```

#### 列出所有欄位

```python
lst = app_train.columns.tolist()
print(lst)
```

#### 截取部分資料

```python
# 取索引100 - 150，SK_ID_CURR與NAME_CONTRACT_TYPE的資料
app_train.loc[100:150,['SK_ID_CURR','NAME_CONTRACT_TYPE']]
```
